```python
# LEAS_DATE의 일자를 추출하기 위해 문자열로 만들어주기 
bicycle_now['tempDATE']= bicycle_now.LEAS_DATE.apply(lambda x:x.split())
```

```python
bicycle_now['LEAS_hour']=bicycle_now.tempDATE.apply(lambda x:x[1].split(':')[0])
bicycle_now['LEAS_min']=bicycle_now.tempDATE.apply(lambda x:x[1].split(':')[1])
bicycle_now['LEAS_sec']=bicycle_now.tempDATE.apply(lambda x:x[1].split(':')[2])
```

```python
#아래 조건문 사용 위해 해당 컬럼을 숫자로 바꿔줌 
bicycle_now["LEAS_hour"] = pd.to_numeric(bicycle_now["LEAS_hour"])
bicycle_now
```

```python
#조건문을 가진 특정 값들만 DF에 보여주기 
bicycle_late2 = bicycle_now[(bicycle_now["LEAS_hour"] >=0) & (bicycle_now["LEAS_hour"] <4 )]
bicycle_late2
```

```python
bicycle_late2.to_csv(r'C:/Users/ASIA_12/00 slee/SBJ_2007_001/전처리_버스/bicycle_late2.csv')
```
