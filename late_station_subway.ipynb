```python
import pandas as pd
import numpy as np
```

```python
leas_csv = pd.read_csv("C:/Users/ASIA_12/00 slee/SBJ_2007_001/b_leas.csv")
```

```python
rtn_csv = pd.read_csv("C:/Users/ASIA_12/00 slee/SBJ_2007_001/b_rtn.csv")
```

```python
late_cycle = pd.read_csv("C:/Users/ASIA_12/00 slee/SBJ_2007_001/전처리_버스/bicycle_late2.csv")
```

## LEASE_STATION 중에서 막차 시간대에 가장 건수가 많은 Top 10

```python
late_cycling_leas = late_cycle.merge(leas_csv, how = 'inner', on = ["LEAS_STATION"])
late_cycling_leas
#This will perform and "inner-join" thereby omitting rows in each dataframe that do not match. 
# Hence, no NaN in either the right or left part of merged dataframe.
```

```python
late_cycling_leas2 = late_cycling_leas.sort_values(by=["정류장 이용 건수"], ascending=False)
late_cycling_leas2
```

```python
late_cycling_leas2.nlargest(50, '정류장 이용 건수')
```

## RETURN_STATION 중에서 막차 시간대에 가장 건수가 많은 Top 10

```python
late_cycling_rtn= late_cycle.merge(rtn_csv, how = 'inner', on = ["RTN_STATION"])
late_cycling_rtn
```

```python
late_cycling_rtn2 = late_cycling_rtn.sort_values(by=["정류장 이용 건수"], ascending=False)
late_cycling_rtn2
```

```python
late_cycling_rtn2.nlargest(10, '정류장 이용 건수')
```
